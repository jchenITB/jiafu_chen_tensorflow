```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

#Datos

x = np.array([189,190,208,227,239,252,257,274,293,308,316])
y = np.array([402,404,412,425,429,436,440,447,458,469,469])

#Crear modelos

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mean_squared_error'
)

betterModel = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=3, input_shape=[1]),
    tf.keras.layers.Dense(units=3),
    tf.keras.layers.Dense(units=1)
])
betterModel.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mean_squared_error'
)
```

```python
#Entrenamiento
modelHistory = model.fit(x, y, epochs=2000, verbose=0)
betterModelHistory = betterModel.fit(x, y, epochs=2000, verbose=0)

#Grafico de la salida
plt.plot(modelHistory.history['loss'], label='Model inicial')
plt.plot(betterModelHistory.history['loss'], label='Modelo con 3 capas')
plt.legend()
plt.show()
```
